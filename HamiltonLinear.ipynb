```python
import time
def hamilton(adj):
    
    
    for i in range(len(adj)):
        adj[i]=[j for j in adj[i] if j!=i]
        adj[i].append(0)
        adj[i].append(len(adj[i])-2)
        adj[i].append(0)
    
    queue=[0]
    adj[0][-1]=1
    
    
    while True:
                       
        if adj[queue[-1]][-3]>adj[queue[-1]][-2]:
            adj[queue[-1]][-3]=0
            adj[queue[-1]][-1]=0
            queue.pop()
            if not queue:
                return "No"
            continue
            
        if adj[queue[-1]][0]==0 and len(queue)==len(adj):
            return "Hamilton",queue
        
        if adj[adj[queue[-1]][adj[queue[-1]][-3]]][-1]==0:
            
            queue.append(adj[queue[-1]][adj[queue[-1]][-3]])
            adj[queue[-2]][-3]+=1
            adj[queue[-1]][-1]=1
        
        else:
            adj[queue[-1]][-3]+=1
                
        if queue[-1]==0:
            if len(queue)==len(adj):
                return "Hamilton",queue
            
```
